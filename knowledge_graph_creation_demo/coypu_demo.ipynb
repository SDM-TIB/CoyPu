```python
#endpoints = ['http://node2.research.tib.eu:11381/sparql', 'https://query.wikidata.org/sparql']
endpoints = {
  'https://labs.tib.eu/sdm/worldbank_endpoint/sparql': {
    'mappings': [
      'mappings/python_mapping_0.ttl',
      'mappings/python_mapping_7.ttl'
    ]
  }
}
create_rdfmts(endpoints, 'config_files/rdfmts.json') #Generating configuration file for DeTrusty

os.system("cat config_files/rdfmts.json") #Printing configuration file for DeTrusty
```

```python
config = ConfigFile('config_files/coypu-demo.json')
query = """
PREFIX p: <http://www.wikidata.org/prop/>
PREFIX ps: <http://www.wikidata.org/prop/statement/>

SELECT DISTINCT ?isoCode ?year ?life_exp WHERE {
    <http://worldbank.org/Country/DEU> a <http://worldbank.org/Country> .
    <http://worldbank.org/Country/DEU> <http://www.w3.org/2002/07/owl#sameAs> ?sameAsCountry .
    <http://worldbank.org/Country/DEU> <http://worldbank.org/hasAnnualIndicatorEntry> ?annualIndicator .
    ?annualIndicator <http://worldbank.org/hasIndicator> <http://worldbank.org/Indicator/SP.DYN.LE00.IN> .
    ?annualIndicator <http://www.w3.org/2002/07/owl#hasValue> ?life_exp .
    ?annualIndicator <http://www.w3.org/2006/time#year> ?year .
    
    ?sameAsCountry p:P298 ?itemIsoCode .
    ?itemIsoCode ps:P298 ?isoCode .
} ORDER BY ?year
"""
query_result = run_query(query, config=config, join_stars_locally=False) #Executing query with DeTrusty

for row in query_result['results']['bindings']:
    print("isoCode: " + row['isoCode']['value'] + " year: " + row['year']['value'] + " Life Expectancy: " + row['life_exp']['value'])
```
